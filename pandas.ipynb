```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame (skip this if already loaded from CSV)
data = {
    'prod name': [
        'LG Mobile',
        'Apple iPad 10.2-inch',
        '34in Ultrawide Monitor',
        'iPhone 11',
        'Bose SoundSport Headphones'
    ],
    'prod cost': [65999, 63000, 75999, 60000, 69999]
}

df = pd.DataFrame(data)

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    df['prod cost'],
    labels=df['prod name'],
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Product Cost Distribution')
plt.axis('equal')
plt.show()
```
